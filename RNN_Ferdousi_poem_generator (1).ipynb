```python
import os
import random

import matplotlib.pyplot as plt
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
import torchtext
from torch.utils.tensorboard import SummaryWriter
from torchtext.legacy.data import Field

############################
########## UTILS ###########
############################
opt_path = "drive/MyDrive/RNN_States/optim_my_checkpoint.pth.tar"
model_path = "drive/MyDrive/RNN_States/model_my_checkpoint.pth.tar"
load_model = True
save_model = True


def save_checkpoint(model, optimizer):
    print("=> Saving checkpoint")
    torch.save(model.state_dict(), model_path)
    torch.save(optimizer.state_dict(), opt_path)


def load_checkpoint(model, optimizer):
    print("=> Loading checkpoint")
    model.load_state_dict(torch.load(model_path))
    optimizer.load_state_dict(torch.load(opt_path))


############################
##### random initializer ###
############################

torch.manual_seed(11747)
random.seed(17757)
np.random.seed(7171757)

############################
## load and tokenize data ##
############################


colab = True
if colab:
    url = 'https://github.com/amirmosio/ferdosi-poem-generator/raw/main/ferdosi.txt'
    file_path = torchtext.legacy.utils.download_from_url(url)
    poem = open(file_path, 'rb').read().decode('utf-8')
    splitter = "\n"
else:
    file_path = 'ferdosi.txt'
    poem = open(file_path, 'rb').read().decode('utf-8')
    splitter = "\r\n"


def tokenize_per(text):
    return [tok for tok in text.replace(splitter, " ").split()]


poems_sentences = np.array(poem.split(splitter))
all_ids = np.array(range(len(poems_sentences) - 1))

test_poems_ids = np.random.choice(len(poems_sentences) - 1, 10)
train_poems_ids = np.delete(all_ids, test_poems_ids)

train_sentences = poems_sentences[train_poems_ids]
test_sentences = poems_sentences[test_poems_ids]

### language
SOS_token = "<sos>"
EOS_token = "<eos>"
persian = Field(tokenize=tokenize_per, lower=True, init_token=SOS_token, eos_token=EOS_token)
persian.build_vocab([tokenize_per(sent) for sent in poems_sentences], max_size=10000, min_freq=2)
SOS_index = persian.vocab[SOS_token]
EOS_index = persian.vocab[EOS_token]

######################
### configuration ####
######################
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(device)

num_epochs = 100
learning_rate = 0.01
batch_size = 64

input_size_encoder = len(persian.vocab)
input_size_decoder = len(persian.vocab)
output_size = len(persian.vocab)
embedding_size = 300
hidden_size = 300
num_layers = 2
enc_dropout = 0.5
dec_dropout = 0.5


######################
####### models #######
######################

class PoemsDataset:
    def __init__(self, data):
        self.data = data
        self.length = len(data)

    @staticmethod
    def convert_index_to_token(array):
        res = []
        for sentence in array:
            token_sentence = []
            for index in sentence:
                token_sentence.append(persian.vocab.itos[index])
            res.append(token_sentence)
        return res

    @staticmethod
    def represent_data(input, output, target):
        for sentence_index in range(len(input)):
            print("___________________________________")
            print("|++: ", " ".join(input[sentence_index]))
            print("|->: ", " ".join(output[sentence_index]))
            print("|=>: ", " ".join(target[sentence_index]))

    @staticmethod
    def convert_sentence_tokens_to_index(array, max_sentence_words=14):
        res = torch.full((array.shape[0], max_sentence_words), EOS_index, dtype=torch.int, device=device)
        for i in range(len(array)):
            for j in range(len(array[i])):
                res[i][j] = persian.vocab[array[i][j]]
        return res

    @staticmethod
    def convert_sentence_tokens_to_vectors(array, max_sentence_words=14):
        res = torch.zeros((array.shape[0], max_sentence_words, len(persian.vocab)), dtype=torch.int, device=device)
        for i in range(array.shape[0]):
            for j in range(array.shape[1]):
                index = array[i][j]
                vec = torch.zeros(len(persian.vocab))
                vec[index] = 1
                res[i][j] = vec
        return res

    def get_batches(self, batch_size=64, max_sentence_words=14):
        for sentence_id in range(0, self.length, batch_size):
            batch = self.data[sentence_id:min(self.length, sentence_id + batch_size)]
            s = np.array([[SOS_token] + tokenize_per(po.split(",")[0]) + [EOS_token] for po in batch])
            s = self.convert_sentence_tokens_to_index(s, max_sentence_words=max_sentence_words)
            t = np.array([[SOS_token] + tokenize_per(po.split(",")[1]) + [EOS_token] for po in batch])
            t = self.convert_sentence_tokens_to_index(t, max_sentence_words=max_sentence_words)
            yield s, t


class Encoder(torch.nn.Module):
    def __init__(self, input_size, embedding_size, enc_hid_dim, dec_hid_dim, layer_number, dropout):
        super().__init__()

        self.embedding = torch.nn.Sequential(
            torch.nn.Embedding(input_size, embedding_size),
            torch.nn.Dropout(dropout)
        )

        self.rnn = torch.nn.LSTM(embedding_size, enc_hid_dim, num_layers=layer_number)

        self.hidden_fc = torch.nn.Sequential(
            torch.nn.Linear(enc_hid_dim, dec_hid_dim),
            torch.nn.ReLU(inplace=True)
        )

        self.fc = torch.nn.Sequential(
            torch.nn.Linear(enc_hid_dim, dec_hid_dim),
            torch.nn.ReLU(inplace=True)
        )

    def forward(self, src):
        src = src.permute(1, 0)
        embedded = self.embedding(src)

        outputs, (hidden, cell) = self.rnn(embedded)

        hidden = self.hidden_fc(hidden)
        cell = self.fc(cell)

        return outputs, hidden, cell


class Decoder(nn.Module):
    def __init__(self, input_size, embedding_size, hidden_size, output_size, num_layers, dropout):
        super(Decoder, self).__init__()

        self.embedding = torch.nn.Sequential(
            torch.nn.Embedding(input_size, embedding_size),
            torch.nn.Dropout(dropout)
        )
        self.attention_weights = torch.nn.Sequential(
            nn.Linear(hidden_size * 3, hidden_size * 2),
            nn.ReLU(),
            nn.Linear(hidden_size * 2, 1),
            nn.ReLU(),
            nn.Softmax(dim=1)
        )
        self.rnn = nn.LSTM(hidden_size + embedding_size, hidden_size, num_layers)
        self.fc = torch.nn.Sequential(
            torch.nn.Linear(hidden_size * 2 + embedding_size, output_size),
            torch.nn.Softmax(dim=2)
        )

    def forward(self, x, lstm_state, hidden, cell):


        embedding = self.embedding(x)

        lstm_state = lstm_state.permute(1, 0, 2) 
        rep = lstm_state.shape[1]
        new_hidden = torch.cat((hidden[0], hidden[1]), dim=1).unsqueeze(1).repeat(1, rep, 1)

        attention = self.attention_weights(torch.cat((new_hidden, lstm_state), dim=2)).permute(0, 2, 1)

        weighted = torch.bmm(attention, lstm_state).permute(1, 0, 2).squeeze(0)
        lstm_input = torch.cat((weighted, embedding), dim=1).unsqueeze(0)
    

        lstm_output, (hidden, cell) = self.rnn(lstm_input, (hidden, cell))

        fc_input = torch.cat((lstm_input, lstm_output), dim=2)

        predictions = self.fc(fc_input)

        return predictions, hidden, cell


class Seq2Seq(nn.Module):
    def __init__(self, encoder, decoder):
        super(Seq2Seq, self).__init__()
        self.encoder = encoder
        self.decoder = decoder

    def forward(self, source, target, teacher_force_ratio=0.5, k=3):

        encoder_outputs, hidden, cell = self.encoder(source)

        batch_size = source.shape[0]
        seq_length = source.shape[1]

        outputs = torch.zeros(batch_size, seq_length, len(persian.vocab)).to(device)
        outputs[:, 0, SOS_index] = torch.ones(batch_size).to(device)

        batch_input_word = torch.tensor([SOS_index] * batch_size, dtype=torch.int, device=device)

        for i in range(1, seq_length):
            output, hidden, cell = self.decoder(batch_input_word, encoder_outputs, hidden, cell)
            output, hidden, cell = output.squeeze(0), hidden, cell

            outputs[:, i] = output

            teacher_force = random.random() < (0.1 if not self.training else teacher_force_ratio)

            batch_input_word = target[:, i] if teacher_force else output.topk(1, dim=-1)[1].squeeze(1)
        if self.training:
            return outputs
        else:
            return outputs.topk(1, dim=-1)[1].squeeze(2)


writer = SummaryWriter(f"runs/loss_plot")

train_dataset = PoemsDataset(train_sentences)
test_dataset = PoemsDataset(test_sentences)

# initializing models
encoder = Encoder(input_size_encoder, embedding_size, hidden_size, hidden_size, num_layers, enc_dropout).to(device)
decoder = Decoder(input_size_decoder, embedding_size, hidden_size, output_size, num_layers, dec_dropout).to(device)
model = Seq2Seq(encoder, decoder).to(device)
# optimizer
optimizer = optim.Adam(model.parameters(), lr=learning_rate)
mse_loss = nn.MSELoss()

if load_model and os.path.isfile(model_path) and os.path.isfile(opt_path):
    load_checkpoint(model, optimizer)

loss_chart_data = []
for epoch in range(num_epochs):
    print(f"[Epoch {epoch} / {num_epochs}]")
    if save_model:
        save_checkpoint(model, optimizer)

    # Eval
    model.eval()
    for test_source, test_target in test_dataset.get_batches(batch_size):
        output = model(test_source, test_target)
        input_tokens = PoemsDataset.convert_index_to_token(test_source)
        out_tokens = PoemsDataset.convert_index_to_token(output)
        target_tokens = PoemsDataset.convert_index_to_token(test_target)
        print(PoemsDataset.represent_data(input_tokens, out_tokens, target_tokens))

    # Train
    model.train()
    for source, target in train_dataset.get_batches(batch_size):
        optimizer.zero_grad()

        output = model(source, target)

        loss = mse_loss(output.float(), PoemsDataset.convert_sentence_tokens_to_vectors(target).float())

        loss.backward()
        loss_chart_data.append(loss.item())
        torch.nn.utils.clip_grad_norm_(model.parameters(), max_norm=1)

        optimizer.step()

        writer.add_scalar("Training loss", loss, global_step=epoch)
    plt.plot(range(len(loss_chart_data)), loss_chart_data)
    plt.show()
```
